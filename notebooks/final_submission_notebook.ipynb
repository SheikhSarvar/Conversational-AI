## 🧠 Phase 3: Fine-Tuning Strategy (LoRA)

We use **Llama 3.2 1B Instruct** as our base model. To enable efficient local and single-GPU training, we use **QLoRA (Quantized Low-Rank Adaptation)**.

**Why LoRA?**
- Drastically reduces trainable parameters (~11M instead of 1B).
- Minimizes VRAM requirements to ~8GB for training.
- Maintains general conversational abilities while supercharging IT-specific logic.
